```python
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib import animation
from matplotlib.animation import PillowWriter
xmin=-50
xmax=50
N=1000
gain=[]
dom1=np.linspace(xmin,xmax,N-2)
posop=np.zeros((N-2,N-2))
for i in range(0,N-2):
  posop[i][i]=dom1[i]
def te(E,i,time):
  te=complex(0+0j)
  te=math.cos(-(E[i]*time))+1j*math.sin(-(E[i]*time))
  return te
def wf(t,ff):
  b=0
  gain=[]
  gain1=[]
  ngg=np.zeros(N-2,dtype=complex)
  ng=np.zeros(N-2,dtype=complex)
  c=1
  dom=np.linspace(xmin,xmax,N)
  m=0.3
  h=1
  j=0
  dt=0.1
  tvals=np.arange(0,t,dt)
  wfm=np.zeros((N-2,len(tvals)))
  dx=(xmax-xmin)/N
  A=0.2
  k=(h*h)/(2*m*dx*dx)
  V=np.zeros(N)
  w=1.0
  trunc=400
  xparr=[]
  guss=np.zeros(N-2)
  for i in range (0,N-2):
    guss[i]=math.exp((-(dom[i]-b)**2)/(2*c*c))
  ng=guss
  for tt in tvals:
    comat=np.zeros(trunc,dtype=complex)
    integral=0
    for i in range (0,len(V)):
      V[i]=0.5*m*w*w*dom[i]*dom[i]-A*dom[i]*math.cos(ff*tt)
    H=np.zeros((N-2,N-2))
    for i in range (0,N-3):
      H[i,i+1]=-k
      H[i+1,i]=-k
      H[i,i]=2*k+V[i]
    H[N-3][N-3]=2*k+V[N-3]
    E,epsi=np.linalg.eig(H)
    idx = E.argsort()[::-1]
    E= E[idx]
    epsi = epsi[:,idx]
    E=np.flip(E)
    epsi1=epsi
    for i in range(0,N-3):
      epsi[:,i]=epsi1[:,N-3-i]
    for i in range (0,trunc):
      comat[i]=np.matmul(epsi[:,i],ng)
    ngg=np.zeros(N-2,dtype=complex)
    for l in range(0,trunc):
      ngg=np.add(ngg,(te(E,l,dt)*comat[l]*epsi[:,l]))
    ngcon=np.conjugate(ngg)
    normfac=np.matmul(ngcon,ngg)
    normfac=np.abs(normfac)
    print(tt,ff)
    ngg=ngg/math.sqrt(normfac)
    ngcon=np.conjugate(ngg)
    expo=np.matmul(posop,ngg)
    expos=np.matmul(ngcon,expo)
    xparr=np.append(xparr,expos)
    print(expos)
    expos=np.abs(expos)
    print(expos)
    gain=np.append(gain,expos)
    maxgain=max(gain)
    npp=np.abs(ngg)
    plt.plot(npp)
    plt.show()
    ng=ngg
    j=j+1
    plt.show()
  plt.plot(tvals,xparr)
  plt.xlabel('t')
  plt.ylabel('<x>')
  plt.title('\u03C9 =')
  plt.show()
  return(maxgain)
gmat=[]
for ttff in [1.0]:
  gmat=np.append(gmat,wf(30,ttff))
  print(gmat)
plt.plot(np.linspace(0.5,1.5,25),gmat)




```
